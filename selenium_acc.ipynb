```python
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from bs4 import BeautifulSoup as bs
import urllib
import glob
import base64
driver = webdriver.Chrome()
driver.get('https://www.acc.go.kr/board/schedule/child')
```

```python
html = bs(driver.page_source,"html")
selected_html = html.select(".list_info") 
titles = []
summaries = []
schedules = []
places = []
costs = []
images = []
baselist= []
```

```python
html = bs(driver.page_source,"html")
selected_html = html.select(".list_info") 

before_titles = html.select('.list_img img')
i=0
for i in range(len(before_titles)):
    titles.append(before_titles[i].get_attribute_list('alt'))

before_summaries = html.select('.list_summary')
for l in before_summaries:
    summaries.append(l.get_attribute_list('data-origin'))

before_schedules = html.select('.list_period2')
for m in before_schedules:
    schedules.append(m.get_attribute_list('data-origin'))

before_places = html.select('.pWidth11per')
for n in before_places:
        places.append(str(n.get_attribute_list('data-origin')).replace(' 지도보기',""))

before_costs = html.select('div.list_price')
for o in before_costs:
    costs.append(o.get_attribute_list('data-origin'))

before_images = html.select('.list_img a img')
homepageUrl = 'https://www.acc.go.kr'
for p in before_images:
    rawUrl = str(p.get_attribute_list('src'))
    rawUrl = rawUrl.replace("['","")
    rawUrl = rawUrl.replace("']","")
    realUrl = homepageUrl+rawUrl
    images.append(realUrl)

for q in range(len(images)):
    urllib.request.urlretrieve(images[q],'acc'+str(q)+'.jpg')

for r in range(len(images)):
    with open('acc'+str(r)+'.jpg','rb') as f:
        baselist.append(base64.b64encode(f.read()))
```
